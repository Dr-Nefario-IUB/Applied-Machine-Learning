```python
train_1=pd.read_csv('train_1.csv')
valid_1=pd.read_csv('valid_1.csv')
train_2=pd.read_csv('train_2.csv')
valid_2=pd.read_csv('valid_2.csv')
test=pd.read_csv('test.csv')



def preprocess(data):
    
    #Replace null values with mean(Imputation)
    data['pdays'].replace(-1,np.nan,inplace=True)
    mean=data['pdays'].mean()
    data['pdays'].replace(np.nan,mean,inplace=True)
    
    #Separate Categorical values from the Data Frame
    categorical=['object']
    categorical_data=data.select_dtypes(categorical)
  
    
    #Label Encoding for features with binary values
    binary_columns=['default','housing','loan','y']
    binary_values_mapping={'yes':1,'no':0}
    for i in binary_columns:
        categorical_data[i]=categorical_data[i].map(binary_values_mapping)
    
    #One hot encoding for nominal categorical values
    nom_cat_col=['job','marital','education','contact','month','poutcome']
    encoded_data=pd.get_dummies(categorical_data[nom_cat_col])


    #encoded_data.head()
    
    #Replace the columns with nominal categorical values, with the dummies generated through one hot encoding 
    categorical_data.drop(nom_cat_col, axis=1,inplace=True)

    for i in encoded_data.columns:
        categorical_data[i]=encoded_data[i].values

    #categorical_data.head()
    
    #Separate the columns with numeric or continous values
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    numeric_columns=data.select_dtypes(numerics)
    #print(numeric_columns.shape)
    #numeric_columns.head()

    #Standardization of numerical column
    
    #print(numeric_columns.columns)
    for i in numeric_columns.columns.values:
        mean=numeric_columns[i].mean()
        std=numeric_columns[i].std()
        standardized_values=[]
        for j in numeric_columns[i].values:
            standardized_values.append((j-mean)/std)
        numeric_columns[i]=standardized_values
    
    #numeric_columns.head()
    
    #Combine the numeric and categorical data
    sanitized_data=pd.DataFrame(numeric_columns.values, columns=numeric_columns.columns)
    #sanitized_data.head()
    for i in categorical_data.columns:
        sanitized_data[i]=categorical_data[i].values

    #sanitized_data.head()

    #print(sanitized_data.values)
    #print(sanitized_data.shape)
    #sanitized_data.info()

    columns=sanitized_data.columns.values
    index=np.where(columns=='y')[0][0]
    l=columns[:index]
    r=columns[index+1:]
    columns=(np.append(l,r))
    columns=np.append(columns,['y'])
    sanitized_data=sanitized_data[columns]
    
    return sanitized_data
    
    
```

Train a logistic regression model with $L_2$ regularization on `train_1.csv` after applying the necessary preprocessing steps of Task 1.
```python
# PROBLEM 1, TASK 2 CODE GOES HERE
import numpy as np
from matplotlib import pyplot as plt
from sklearn.metrics import accuracy_score
class Logistic_Regression_Model:
    def __init__(self, learning_rate,regularization_strength, iterations, coef=None):
        self.learning_rate=learning_rate
        self.regularization_strength=regularization_strength
        self.iterations=iterations
        self.coef=coef
    
    def fit_model(self,feature_data,target):
            coefficiets=np.zeros(feature_data.shape[1])
            
            ones=np.ones(len(feature_data))
            model_cost=[]
            for i in range(0,self.iterations):
                output_vector=[]
                
                for x in range (0,len(feature_data)):
                    linear_hypothesis=coefficiets.dot(feature_data[x])
                    logistic_hypothesis=1/(1+np.exp(-linear_hypothesis))
                    output_vector.append(logistic_hypothesis)
                
                cost_function=(-1/len(feature_data))*((target.dot(np.log(output_vector)))+((ones-target)).dot(np.log(ones-output_vector)))+(self.regularization_strength/(2*len(feature_data)))*(coefficiets.dot(coefficiets.transpose())) 
                #print(cost_function)
                model_cost.append(cost_function)
                
                error=output_vector-target
                
                #Gradient descent
                coefficiets[0]=coefficiets[0]-((self.learning_rate/len(feature_data))*(error.dot(feature_data.transpose()[0])))
                coefficiets[1:]=(1-((self.learning_rate*self.regularization_strength)/len(feature_data)))*coefficiets[1:]-(self.learning_rate/len(feature_data))*(feature_data.transpose()[1:].dot(error))
            
            #plt.plot(model_cost)
            self.coef=coefficiets
    
    def predict(self,feature_data):
        output_vector=[]
        for x in range (0,len(feature_data)):
            linear_hypothesis=self.coef.dot(feature_data[x])
            logistic_hypothesis=1/(1+np.exp(-linear_hypothesis))
            output_vector.append(logistic_hypothesis)
        return output_vector
    def classifier(self,predicted_values):
        classified_values=[]
        for i in predicted_values:
            if i>=0.5:
                classified_values.append(1)
            else:
                classified_values.append(0)
        return classified_values
    def validation(self,validation_feature_data,validation_target_data,train_feature_data,train_target_data):
        
        lambda_values=[i for i in range(0,10)] 
        
        training_accuracy_vector=[]
        validation_accuracy_vector=[]
        for i in lambda_values:
            self.regularization_strength=i
            self.fit_model(train_feature_data,train_target_data)
            y_pred_training=self.predict(train_feature_data)
            y_pred_validation=self.predict(validation_feature_data)
            training_accuracy=accuracy_score(train_target_data,self.classifier(y_pred_training))
            validation_accuracy=accuracy_score(validation_target_data,self.classifier(y_pred_validation))
            training_accuracy_vector.append(training_accuracy)
            validation_accuracy_vector.append(validation_accuracy)
        
        
        print("The best accuracy on training data "+str(max(training_accuracy_vector)))
        print("The best accuracy on validation data "+str(max(validation_accuracy_vector)))
        print("The best value of lambda- "+str(lambda_values[validation_accuracy_vector.index(max(validation_accuracy_vector))]))
        plt.plot(lambda_values,training_accuracy_vector,"b",label="training accuracy")
        plt.plot(lambda_values,validation_accuracy_vector,"r",label="validation accuracy")
        plt.xlabel("Lambda Î»")
        plt.ylabel("Accuracy")
        plt.legend(loc = "upper right")
            
            
        
        
        
            
        
```

```python
import numpy as np
from scipy.special import legendre
from matplotlib import pyplot as plt

def get_coefficients(degree):
    """
    The coefficients returned are in the order w_0, w_1, ... , w_{degree}
    """
    return legendre(degree).coefficients[::-1]


def visualize_function(degree, low, high,samples):
    w = get_coefficients(degree)
    print(w)
    x=np.linspace(low,high,samples)
    #print(len(x))
    #print(x)
    y=0
    for i in range(0,degree+1):
        y+=w[i]*(x**i)
    #print(y)
    plt.plot(x,y)
    plt.xlabel('domain')
    plt.ylabel('f(x)')
## TEST CASES (The following lines should plot a function)

visualize_function(3, -1, 1,20)
visualize_function(4, -1, 1,20)
visualize_function(6, -1, 1,20)
visualize_function(10, -1, 1,20)
```

### Task 2 (5 points)

Write a function in python that randomly generates a dataset (training data + test data). The function should take the degree of target polynomial as input along with the number of points that should be generated for both the training and the test sets. The polynomial will be represented by the vector of coefficients. The code to calculate these coefficients has been provided. The function should also add random normal noise to the target -- i.e. the noise should be drawn from a normal distribution with mean $\mu=0$ and standard deviation = $\sigma^2$. Thus: 
$$y=\sum_{i=0}^d w_ix^i + \mathcal{N}(0, \sigma^2)$$



```python
def generate_dataset(degree, n_train, n_test, sigma):
    """
    - n_train - Number of training samples
    - n_test - Number of test samples
    - X_train is an array of size (n_train $\times$ 1)
    - y_train is an array of size (n_train $\times$ 1)
    - X_test is an array of size (n_test $\times$ 1)
    - y_test is an array of size (n_test $\times$ 1)
    """
    w = get_coefficients(degree)
    X_train = np.linspace(-1,1,n_train)
    X_train=X_train.tolist() 
    X_test = np.linspace(-1,1,n_test)
    X_test=X_test.tolist()
    y_train = []
    y_test = []
    for i in range(0,n_train):
        sum=0
        for d in range(0,degree+1):
            sum+=w[d]*(X_train[i]**d)
        y_train.append(sum)
    
    for i in range(0,n_test):
        sum=0
        for d in range(0,degree+1):
            sum+=w[d]*(X_test[i]**d)
        y_test.append(sum)
    
    Noise_train = np.random.normal(0,sigma,len(y_train))
    Noise_test = np.random.normal(0,sigma,len(y_test))
    y_train+=Noise_train
    y_test+=Noise_test
    
    return X_train, y_train, X_test, y_test


#a,b,c,d=generate_dataset(3,50,100,1)
```

```python
from matplotlib import pyplot as plt
degree=int(input('Enter the degree of the polynomial: '))
sigma=float(input('Enter the deviation in noise'))
n_train=int(input('Enter the number of training samples'))
n_test=int(input('Enter the number of test samples'))
X_train,y_train,X_test,y_test=generate_dataset(degree,n_train,n_test,sigma)
print(X_train)
print(y_train)
print(X_test)
print(y_test)

low=min(X_train)
high=max(X_train)

visualize_function(degree,low,high,n_train)
plt.scatter(X_train,y_train,color='r')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error,r2_score
import numpy as np
import pandas as pd
poly_features_A=PolynomialFeatures(degree=2)
A_X_train=poly_features_A.fit_transform(np.asarray(X_train).reshape(-1,1))
modelA=LinearRegression()
modelA.fit(A_X_train,y_train)
y_pred_A=modelA.predict(A_X_train)

poly_features_B=PolynomialFeatures(degree=10)
B_X_train=poly_features_B.fit_transform(np.asarray(X_train).reshape(-1,1))
modelB=LinearRegression()
modelB.fit(B_X_train,y_train)
y_pred_B=modelB.predict(B_X_train)

training_accuracy_A=np.sqrt(mean_squared_error(y_train,y_pred_A))
training_accuracy_B=np.sqrt(mean_squared_error(y_train,y_pred_B))
r2_modelA=r2_score(y_train,y_pred_A)
r2_modelB=r2_score(y_train,y_pred_B)

print("A: "+str(r2_modelA))
print("B: "+str(r2_modelB))


print(training_accuracy_A)
print(training_accuracy_B)

A_X_test=poly_features_A.fit_transform(np.asarray(X_test).reshape(-1,1))
y_pred_A_test=modelA.predict(A_X_test)
B_X_test=poly_features_B.fit_transform(np.asarray(X_test).reshape(-1,1))
y_pred_B_test=modelB.predict(B_X_test)

test_accuracy_A=np.sqrt(mean_squared_error(y_test,y_pred_A_test))
test_accuracy_B=np.sqrt(mean_squared_error(y_test,y_pred_B_test))
r2_testA=r2_score(y_test,y_pred_A_test)
r2_testB=r2_score(y_test,y_pred_B_test)

print("A: "+str(r2_testA))
print("B: "+str(r2_testB))

print(test_accuracy_A)
print(test_accuracy_B)


#Tabular formation of mertrices

metrics={"RMSE_training":[training_accuracy_A,training_accuracy_B],"RMSE_test":[test_accuracy_A,test_accuracy_B],"R2_Score_training":[r2_modelA,r2_modelB],"R2_Score_test":[r2_testA,r2_testB]}
metrics=pd.DataFrame(data=metrics,index=["Model A","Model B"])

metrics.head(len(metrics.values))
```

While the performance of the model A decreased during testing, the performance didn't decrease by a substantial amount, like model B(in terms of percentage).<br>
The performance of model A was expected. Since the training and test datasets were generated using a polynomial of 10th degree, fitting the model with a 2nd degree polynomial, and training it with a meager data set of 20 samples would lead to a significant error in prediction. <br>

The performance of model B was also expected.
